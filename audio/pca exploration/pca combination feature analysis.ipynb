```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('C:/Users/Administrator/Downloads/pca grid feature scores.csv')
```

```python
df
```

```python
print(df['Accuracy'].max())
print(df['Precision'].max())
print(df['Recall'].max())
```

### ACCURACY ANALYSIS:

<font color = "black">__Max Accuracy obtained is 0.6705__</font>

```python
acc_df=df.sort_values(by=['Accuracy'], ascending=False).head(15)
acc_df
```

```python
l=['pho1','pho2', 'pho3', 'pho4', 'pro1', 'pro2', 'pro3', 'pro4', 'art1', 'art2', 'art3', 'art4']

count=[]
lst=acc_df['Feature'].to_list()
for i in l:
    count.append(sum(x.count(i) for x in lst))
```

```python
count_df=pd.DataFrame(l,columns=['Feature'])
count_df['count']=count
count_df.sort_values(by='count',ascending=False)
```

## PRECISION ANALYSIS:

<font color = "black">__REF: Max Precision obtained is 0.3578__</font>

```python
pr_df=df[df['Precision']>0.33].sort_values(by=['Precision'],ascending=False)
pr_df
```

```python
count_pr=[]
lst=pr_df['Feature'].to_list()
for i in l:
    count_pr.append(sum(x.count(i) for x in lst))
```

```python
count_pr_df=pd.DataFrame(l,columns=['Feature'])
count_pr_df['count']=count_pr
count_pr_df.sort_values(by='count',ascending=False)
```

## RECALL ANALYSIS:

<font color = "black">__REF: Max Recall obtained is 0.74468__</font>

```python
re_df=df[df['Recall']>0.70].sort_values(by=['Recall'],ascending=False)
re_df
```

```python
count_re=[]
lst=re_df['Feature'].to_list()
for i in l:
    count_re.append(sum(x.count(i) for x in lst))
```

```python
count_re_df=pd.DataFrame(l,columns=['Feature'])
count_re_df['count']=count_re
count_re_df.sort_values(by='count',ascending=False)
```
